  8. Read Section 6. What is the "chicken-and-egg" problem with focusing on tidy data? What does Wickham hope happens in the future with further work on the subject of data wrangling?

2. Categorical variable: For the ./data/sharks.csv data covered in the lecture, clean the "Type" variable as well as you can, and explain the choices you make.

```python
df = pd.read_csv('/content/wrangling/assignment/data/sharks.csv', low_memory=False)
```

```python
df['Type'].value_counts()
```

This data is messy because 'Boat', 'Boating', and 'Boatomg', and 'Sea Disaster' are all able to fall under 'Watercraft', so we can replace these values to 'Watercraft'

```python
type = df['Type']

type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft')
type.value_counts()
```

'Unconfirmed', 'Invalid', 'Questionable', 'Unverified', and 'Under investigation' are all also inconclusive values, so we can get rid of them by making them nan

```python
type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan)
type.value_counts()

df['Type'] = type
del type

df['Type'].value_counts()
```

This tells us that sharks are mosy likely to attack when Unprovoked

4. Missing values, not at random: For the pretrial data covered in the lecture, clean the ImposedSentenceAllChargeInContactEvent variable as well as you can, and explain the choices you make. (Hint: Look at the SentenceTypeAllChargesAtConvictionInContactEvent variable.)

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']


length = pd.to_numeric(length,errors='coerce')
length_NA = length.isnull()
print( np.sum(length_NA),'\n')

print( pd.crosstab(length_NA, type), '\n')

length = length.mask( type == 4, 0)
length = length.mask( type == 9, np.nan)
# I mask the values of 4 and 9 with 0 and na

length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n')
# Then I identify the nan values after masking

df['ImposedSentenceAllChargeInContactEvent'] = length
del length, type
```
